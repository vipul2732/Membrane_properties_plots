```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('Density.xlsx')
```

```python
df.head()
```

```python
df1 =pd.melt(df, id_vars=['Distance'], value_vars=['CHL','POPC','POPE','SSM','PSM','Water','Total'], var_name='groups', value_name="Density(Kg/m\u00b3)")
```

```python
sns.set( font_scale=2.2, style='white', font='sans-serif')
g =sns.relplot( x='Distance',y="Density(Kg/m\u00b3)", kind='line', data=df1, height=8, aspect=8/6, linewidth=2, hue='groups').set(xlabel='Distance(nm)', ylabel='Density(kg/m\u00b3)')
g.savefig('Density.jpg', dpi=400)
```
